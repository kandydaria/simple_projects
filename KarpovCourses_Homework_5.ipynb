Мини-анализ данных о транзакциях за период с 01.12.2010 по 12.09.2011. Найдем и очистим датафрейм от дубликатов, удалим из них отмененные транзакции, посмотрим данные о транзакциях для пользователей конкретной страны, совершивших наибольшее кол-во транзакций (более 80%-перцентиля), посмотрим, какие продукты добавляли в корзину чаще всего, и найдем топ-5 транзакций по сумме заказа.

**Описание данных:**

InvoiceNo – номер транзакции     
StockCode – код товара     
Description – описание товара    
Quantity – количество единиц товара, добавленных в заказ   
InvoiceDate – дата транзакции    
UnitPrice – цена за единицу товара   
CustomerID – id клиента  
Country – страна, где проживает клиент   

```python
import pandas as pd
```

```python
retail = pd.read_csv('https://stepik.org/media/attachments/lesson/361623/data.csv.zip',compression='zip',encoding='ISO-8859-1')
retail
```

```python
retail_columns = retail.columns
retail_columns
```

```python
# Дубликаты в данных:

retail.loc[retail.duplicated()]

# или
# retail[retail.duplicated()]
```

```python
# Очистим данные от дубликатов:

retail = retail.drop_duplicates()
retail
```

```python
# Данные содержат в себе записи как и об успешных транзакциях, так и об отмененных. Если пользователь отменил заказ, 
# в начале номера транзакции (InvoiceNo) ставится C (canceled). Всего таких наблюдений:

retail.InvoiceNo.str.startswith('C').sum()
```

```python
retail = retail.query('Quantity > 0')
```

```python
retail
```

```python
# Посчитаем число заказов для каждого пользователя (CustomerID) из Германии (Germany). Оставьте только тех, кто 
# совершил более N транзакций (InvoiceNo), где N – 80% процентиль. Запишем полученные id пользователей в germany_top 

quantile_08 = retail.query('Country == "Germany"') \
        .groupby('CustomerID') \
        .agg({'InvoiceNo':pd.Series.nunique}) \
        .InvoiceNo.quantile(q=0.8)
quantile_08
```

```python
germany_top = retail.query('Country == "Germany"') \
                    .groupby('CustomerID', as_index = False) \
                    .agg({'InvoiceNo':pd.Series.nunique}) \
                    .query('InvoiceNo > @quantile_08') \
                    .CustomerID
```

```python
# Данные по интересующим нас юзерам(из Германии, совершившие более N транзакций):

top_retail_germany = retail[retail.CustomerID.isin(germany_top)]

# или 
# top_retail_germany = retail.query('CustomerID in @germany_top')

top_retail_germany
```

```python
# Какой из продуктов добавляли в корзину чаще всего, кроме POST? 

top_retail_germany.query('StockCode != "POST"') \
                    .groupby('StockCode') \
                    .agg({'Quantity' : 'count'}) \
                    .sort_values('Quantity', ascending = False)   
# .idxmax()
```

```python
top_retail_germany.StockCode.value_counts()
```

```python
# Добавим колонку с суммой покупки:

retail = retail.assign(Revenue = retail.Quantity * retail.UnitPrice)
retail
```

```python
# Топ-5 транзакций по сумме заказа:

retail.groupby('InvoiceNo',as_index = False) \
    .agg({'Revenue':'sum'}) \
    .sort_values('Revenue',ascending = False) \
    .head() \
    .InvoiceNo.str.cat(sep=', ')
```
