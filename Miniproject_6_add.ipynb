Работа со строковыми данными формата .tsv. Данные доступны по ссылке https://stepik.org/media/attachments/course/72204/gff.tsv     


```python
import pandas as pd
```

```python
# Найдем, какое значение атрибута Parent самое частое, для этого запишем в список все встречаемые атрибутивы

lst = []
with open('../Downloads/gff.tsv','r') as f:
    for line in f:
        if 'Parent=' in line:
            a = line.split('\t')[-1].rstrip().split(';')
            for i in a:
                if i.startswith('Parent=') :
                    lst.append(i.lstrip('Parent=').rstrip(','))
print(lst)
             
```

```python
# Сформируем датафрейм из полученного списка:

data = pd.DataFrame(lst)
data
```

```python
# Кол-во уникальных значений атрибутива:

data[0].value_counts()
```

```python
# Теперь найдем самое частое значение в колонке type

lst=[]
with open('../Downloads/gff.tsv','r') as f:
    for line in f:
        if 'Parent=' in line:
            a = line.split('\t')[2]
            lst.append(a)
```

```python
data = pd.DataFrame(lst)
data
```
