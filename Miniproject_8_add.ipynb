В этой работе нет анализа данных - здесь нам нужно привести данные из Excel в нормальный\читабельный вид - собрать данные из разных табличек в один датафрейм.

```python
import pandas as pd
```

```python
data = pd.read_excel('https://stepik.org/media/attachments/course/72204/tables.xlsx')
data
```

```python
data1 = data.iloc[22:27]
data1.columns = ['company41', 'company51', 'company50', 'company76',
       'company47', 'company27', 'company48','none']
data1 = data1.drop('none', axis=1)
data1
```

```python
data = data.drop(index=[22,23,24,25,26])
data = pd.concat([data,data1])
data = data.drop_duplicates(keep = False)
data = data.drop(columns="Unnamed: 0").reset_index(drop=True)
data
```
